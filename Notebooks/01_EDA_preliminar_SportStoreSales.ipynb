```python
df_sales_0.columns = df_sales_0.columns.str.replace(r"\s+", "_", regex=True).str.strip()
df_product_0.columns =df_product_0.columns.str.replace(r"\s+", "_", regex=True).str.strip()
```

```python
import re
```

```python
df_sales_0.columns = [re.sub(r'_+', '_', c) for c in df_sales_0.columns]
df_product_0.columns = [re.sub(r'_+', '_', c) for c in df_product_0.columns]
```

```python
df_sales_product_0 = df_sales_0.merge(df_product_0, on = 'Product_ID', how = "left")
df_sales_product_0.head(3)
```

```python
df_sales_product_0.shape
print(f"Tenemos un total de {df_sales_product_0.shape[0]} filas y de {df_sales_product_0.shape[1]} columnas")
```

```python
df_sales_product_0.info()
```

```python
for col in ['Sale_Price_EUR','Cost_Price_EUR','Discount_percent']:
    df_sales_product_0[col] = (df_sales_product_0[col].astype(str).str.replace(',', '.', regex=False).astype(float))
```
