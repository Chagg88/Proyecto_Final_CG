```python
df_sales_2023_2024.info()
```

```python
df_sales_2023_2024['Grosss_Sale'] = df_sales_2023_2024['Sale_Price_EUR'] * df_sales_2023_2024['Quantity_Sold'] 
df_sales_2023_2024.head(5)
```

```python
df_sales_2023_2024['Total_Discount'] = df_sales_2023_2024['Sale_Price_EUR'] * df_sales_2023_2024['Discount_percent'] 
df_sales_2023_2024.head(6)
```

```python
df_sales_2023_2024['Net_Sales'] = df_sales_2023_2024['Gross_Sales'] - df_sales_2023_2024['Total_Discount'] 
df_sales_2023_2024.head(6)
```

```python
df_sales_2023_2024['Total_Cost'] = df_sales_2023_2024['Cost_Price_EUR'] * df_sales_2023_2024['Quantity_Sold'] 
df_sales_2023_2024.head(6)
```

```python
df_sales_2023_2024['Total_Profit_EUR'] = df_sales_2023_2024['Net_Sales'] + df_sales_2023_2024['Total_Cost'] 
df_sales_2023_2024.head(6)
```

```python
df_sales_2023_2024['Profit_Percent'] = df_sales_2023_2024['Total_Profit_EUR'] / df_sales_2023_2024['Net_Sales'] 
df_sales_2023_2024.head(8)
```
